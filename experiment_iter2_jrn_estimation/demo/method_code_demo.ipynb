# JRN Estimation & Multiplicative Compounding on rel-stack

**Join Reproduction Number (JRN)** quantifies how much a foreign-key join improves
a downstream prediction task. This notebook demonstrates:

1. **Phase 1 — Individual JRN Estimation**: For each of 11 FK joins × 3 entity tasks
   on the rel-stack dataset, we measure how aggregated child-table features improve
   an MLP probe's metric (AUROC for classification, neg-MAE for regression).
   `JRN = M_join / M_base` (corrected so JRN > 1 always means improvement).

2. **Phase 3 — Multiplicative Compounding**: For 7 multi-hop join chains (2- and 3-hop),
   we test whether `JRN_chain ≈ ∏ JRN_individual` — i.e., whether JRN composes
   multiplicatively across hops (R² = 0.83).

3. **Baseline Comparison**: A fan-out heuristic predicts JRN from join statistics alone
   (coverage × log fan-out), without any model training.

All results are pre-computed using lightweight MLP probes (hidden_dim=32, 2 layers,
10 epochs, 3 seeds).
