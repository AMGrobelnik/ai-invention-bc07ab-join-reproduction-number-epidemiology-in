# RelBench rel-f1 Dataset Preparation for JRN Estimation

This notebook demonstrates the **RelBench rel-f1 (Formula 1) relational database** preparation pipeline for Join Reproduction Number (JRN) estimation.

**What it does:**
- Loads the raw RelBench rel-f1 data (tables, FK metadata, prediction tasks)
- Transforms each record type into a unified example format with `input`/`output` strings and `metadata_*` fields
- Produces a single dataset (`rel-f1`) containing table rows, FK join metadata entries, and prediction task samples

**Source:** Stanford RelBench v2.1.1, [arXiv:2407.20060](https://arxiv.org/abs/2407.20060), CC-BY-4.0
