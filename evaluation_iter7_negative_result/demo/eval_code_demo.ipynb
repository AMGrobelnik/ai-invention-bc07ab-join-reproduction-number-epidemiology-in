# Negative-Result Diagnostic: Why Threshold and Multiplicative Compounding Predictions Failed

**Systematic post-hoc diagnostic** of two hypothesis failures:
1. **Inverted-U threshold prediction** for aggregation sensitivity
2. **Multiplicative compounding model** for chain JRN

Computes 14 metric groups (M1-M14) across 3 dependency experiments:
- **exp_id1_it3**: GBM probes (65 join-task pairs)
- **exp_id1_it2**: MLP probes (51 join-task pairs)
- **exp_id2_it6**: Chain compounding (44 chains)

Key findings: aggregation sensitivity exists (mean CoV=0.109) but is unrelated to JRN (ρ=-0.019); log-linear β coefficients are strongly sub-multiplicative (GBM β=0.17), explaining why multiplicative compounding fails while log-linear succeeds.
