# JRN Probe Cost-Efficiency Replication on rel-stack

**Join Reproduction Number (JRN)** analysis on RelBench rel-stack dataset (Stack Exchange, 7 tables, 11 FK joins, 4.2M rows).

This notebook demonstrates:
1. **Convergence analysis**: Do cheap probes produce the same JRN rankings as expensive full-budget models?
2. **Cost-efficiency comparison**: JRN-based selection vs greedy forward selection vs random baselines across 3 tasks
3. **Cross-dataset generalization**: Does JRN ranking quality generalize from rel-f1 to rel-stack?

Key finding: JRN-based selection achieves 98.8–100% of greedy performance. With only 5–6 direct joins per task, greedy is already cheap, yielding speedup <1× — validating the theoretical breakeven at J≈8 joins.
