# GBM Probe JRN Estimation on rel-stack

**Join Reproduction Number (JRN)** quantifies how much a foreign-key join improves predictive performance. This notebook analyses pre-computed JRN results from LightGBM probes across 11 FK joins in the StackOverflow-based `rel-stack` relational benchmark, covering 3 entity-level tasks and 5 aggregation strategies.

**What this demo covers:**
1. **JRN Matrix** — JRN values for each (join, task, aggregation) triple
2. **Probe-to-Full Spearman ρ** — correlation between lightweight probe JRN and full-model JRN
3. **Aggregation Sensitivity** — quadratic fit of sensitivity vs JRN (inverted-U hypothesis)
4. **Multiplicative Compounding** — whether multi-hop chain JRN ≈ product of single-hop JRNs
