# RelBench rel-hm Dataset Preparation for JRN Estimation

This notebook demonstrates the **RelBench rel-hm (H&M)** relational dataset prepared for Join Reproduction Number (JRN) estimation experiments.

**What this artifact does:**
- Loads the H&M Personalized Fashion Recommendations dataset (3 tables: customers, articles, transactions)
- Computes full join statistics across 2 FK joins with extreme cardinality variation
- Creates a stratified subsample preserving natural fan-out distribution
- Produces examples with time-based train/val/test splits for 3 prediction tasks

**Dataset overview:**
| Table | Rows (full) | Rows (subsample) | Task | Type |
|---|---|---|---|---|
| customers | 1,371,980 | 10,836 | user-churn | classification |
| articles | 105,542 | 47,752 | item-sales | regression |
| transactions | 31,788,324 | 249,172 | price | regression |

**FK joins:** `transactions.customer_id → customers` (fan-out 1–1,895), `transactions.article_id → articles` (fan-out 1–50,287)
