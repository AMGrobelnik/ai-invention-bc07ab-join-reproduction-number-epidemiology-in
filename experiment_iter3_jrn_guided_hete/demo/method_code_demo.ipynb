# JRN-Guided Heterogeneous Architecture: GBM Probe Estimation & 5-Config Comparison

**Join Reproduction Number (JRN)** estimates how much each FK join in a relational database improves prediction for a given task. This notebook:

1. **Loads** Formula 1 relational data (races, drivers, standings, qualifying, results, etc.)
2. **Engineers features** from entity tables and FK joins using aggregation strategies
3. **Estimates JRN** for each of 13 FK joins using LightGBM probes (base metric vs. base+join metric)
4. **Classifies joins** as HIGH (JRN > 1.15), CRITICAL (0.85-1.15), or LOW (< 0.85)
5. **Compares 5 configurations**: JRN-guided (heterogeneous), uniform-mean, uniform-rich, top-K, and oracle
6. **Visualizes** the JRN matrix and configuration comparison results
