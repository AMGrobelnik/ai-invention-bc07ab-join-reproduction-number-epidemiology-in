# JRN Multiplicative Compounding: 2×2 Probe Grid on rel-f1

**Join Reproduction Number (JRN)** measures how much predictive signal a foreign-key join adds to a machine-learning probe. This notebook:

1. Measures **individual JRN** for all 13 FK joins in the Formula 1 relational database using both MLP and GBM probes
2. Measures **enrichment JRN** for join-enrichment pairs (adding parent features to child rows)
3. Measures **chain JRN** for multi-hop chains across driver, circuit, and constructor prediction tasks
4. Compares **4 compounding models** (multiplicative, additive, bottleneck, log-linear) to predict chain JRN from individual hop JRNs

**Key finding:** Log-linear compounding best predicts chain JRN, while naive multiplicative/additive models have negative R².
