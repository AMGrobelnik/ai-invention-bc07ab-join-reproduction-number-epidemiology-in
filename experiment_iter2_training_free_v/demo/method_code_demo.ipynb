# Training-Free vs Probe-Based JRN: Diagnostic Comparison on rel-f1

**Compares 5 training-free statistical proxies** (fanout, Pearson correlation, mutual information, conditional entropy reduction, homophily) **against probe-based JRN** (GBM and MLP) for predicting join utility across FK joins in the rel-f1 (Formula 1) schema.

Evaluates (join, task) pairs using relbench driver tasks and self-supervised tasks for races/constructors/circuits. Outputs per-join results, Spearman correlation matrix, Steiger's Z-test, and per-task analysis.

**Key findings:** GBM probe JRN achieves high Spearman ρ with ground truth, while conditional entropy reduction is the best training-free alternative.
